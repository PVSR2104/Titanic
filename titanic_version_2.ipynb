```python
#reading data
import pandas as pd
data = pd.read_csv(r'C:\Users\SpaceTime\Desktop\code\titanic\train.csv')
test = pd.read_csv(r'C:\Users\SpaceTime\Desktop\code\titanic\test.csv')
```

```python
data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#encodind sex and embarked from categorical data numeric data
data[['Sex','Embarked']] = OrdinalEncoder().fit(data[['Sex','Embarked']]).transform(data[['Sex','Embarked']])
test[['Sex','Embarked']] = OrdinalEncoder().fit(test[['Sex','Embarked']]).transform(test[['Sex','Embarked']])
#scalling the values of age and fare to 0 to 1 to decrease bias over these two features
from sklearn.preprocessing import MinMaxScaler
data[['Age','Fare']] = MinMaxScaler().fit(data[['Age','Fare']]).transform(data[['Age','Fare']])
test[['Age','Fare']] = MinMaxScaler().fit(test[['Age','Fare']]).transform(test[['Age','Fare']])
```

```python
#filling missing values with mean value ,Dont wheater it is good or not
data['Age'] = data['Age'].fillna(data['Age'].median())
test['age'] =test['Age'].fillna(test['Age'].median())
```

```python
#data = data.drop(['Ticket','Cabin','Name'],axis=1)
data=  data.dropna()
data.isnull().any()
```

```python
#spearating data for training
train_set = data[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
train_label = data['Survived']
test_set =test[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
len(test_set)
```

```python
from sklearn.utils import shuffle
train_set,train_label = shuffle(train_set,train_label)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Activation,Dense
from keras.metrics import binary_crossentropy
```

```python
model = Sequential([
    Dense(units = 32,input_shape =(7,) ,activation ='relu' ),
    Dense(units = 16,activation ='relu' ),
    Dense(units = 1,activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_set,train_label,epochs=1000,batch_size=10,validation_split =0.25,shuffle =True)
```

```python
prediction = model.predict(test_set)
```

```python
prediction = [1 if i>0.5 else 0 for i in prediction]
```
